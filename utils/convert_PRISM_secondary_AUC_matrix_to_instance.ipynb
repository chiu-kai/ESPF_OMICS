```python
import pandas as pd
inputfilename ="../../data/no_Imputation_PRISM_Repurposing_Secondary_Screen_data/Drug_sensitivity_AUC_(PRISM_Repurposing_Secondary_Screen)_subsetted_NOrepeat.csv"
AUC_df = pd.read_csv(inputfilename, sep=',', index_col=0)
```

```python
AUC_df.reset_index(inplace=True)
AUC_df
```

```python
# Melt the DataFrame to long format
AUC_long_df = pd.melt(AUC_df, id_vars=['CCL'], var_name='Drug', value_name='Value')
AUC_long_df
```

```python
# Drop rows with missing values
AUC_long_df = AUC_long_df.dropna()
AUC_long_df
```

```python
# Output the long_df DataFrame to a CSV file
AUC_long_df.to_csv('Drug_sensitivity_AUC_(PRISM_Repurposing_Secondary_Screen)_subsetted_NOrepeat_instance_format.csv', index=False)
```
