```python
import pubchempy as pcp

def get_smiles_from_pubchem(drug_names):
    if isinstance(drug_names, str):  # 
        drug_names = [drug_names]

    smiles_dict = {}
    for Dname in drug_names:
        try:
            compound = pcp.get_compounds(Dname, "name")
            smiles_dict[Dname] = compound[0].isomeric_smiles if compound else "Not found"
        except:
            smiles_dict[Dname] = "Not found"
    
    return smiles_dict


smiles_results = get_smiles_from_pubchem(drug_name_ls)
```
