```python
def compare_lists(list1, list2):
    diff_bit_indices=[]
    for i in range(len(list1)):
        if list1[i] != list2[i]:
            diff_bit_indices.append(i)
    if diff_bit_indices:
        print("mismatch")
        print(f'total {len(diff_bit_indices)} bits are different at {diff_bit_indices}')
        return False
    else:
        print("match")
        return True
    
def all_strings_same(ls):
    first_string = ls[0].split(',')
    for string in ls[1:]:
        diff_bit_indices=[]
        string =string.split(',')
        diff_bit_indices = [i for i in range(len(string)) if string[i] != first_string[i]]
        if diff_bit_indices:
            print("mismatch")
            print(f'total {len(diff_bit_indices)} bits are different at {diff_bit_indices}')
            return False  
    print("match")
    return True
```

```python
def rank(data):
    """  """
    sorted_data = sorted((value, index) for index, value in enumerate(data))
    ranks = [0] * len(data)
    for rank, (value, index) in enumerate(sorted_data, start=1):
        ranks[index] = rank
    return ranks

def spearman_correlation(x, y):
    # 
    rank_x = rank(x)
    rank_y = rank(y)
    print(rank_x)
    print(rank_y)
    # 
    d_squared = sum((rx - ry) ** 2 for rx, ry in zip(rank_x, rank_y))
    
    n = len(x)
    #  Spearman 
    if n == 0:
        return 0
    else:
        return 1 - (6 * d_squared) / (n * (n ** 2 - 1))

# 
x = [2, 12, 3, 4, 5]
y = [2, 4, 5, 4, 1000000000000000000000]

#  Spearman 
print(spearman_correlation(x, y))
```

```python
def pearson_correlation(x, y):
    # 
    mean_x = sum(x) / len(x)
    mean_y = sum(y) / len(y)

    # 
    numerator = sum((xi - mean_x) * (yi - mean_y) for xi, yi in zip(x, y))
    denominator_x = sum((xi - mean_x) ** 2 for xi in x)
    denominator_y = sum((yi - mean_y) ** 2 for yi in y)
    denominator = (denominator_x * denominator_y) ** 0.5

    #  Pearson 
    if denominator == 0:
        return 0
    else:
        return numerator / denominator

# 
x = [1, 2, 3, 4, 5]
y = [0.00001, 4, 5, 4, 100000000]

#  Pearson 
print(pearson_correlation(x, y))
```
